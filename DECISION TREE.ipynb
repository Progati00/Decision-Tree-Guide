# Theoretical Questions:





## 1. What is a Decision Tree, and how does it work?

#### A Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. It works by splitting the dataset into subsets based on the value of input features, creating a tree-like structure of decisions. Each internal node represents a decision based on a feature, each branch represents the outcome of that decision, and each leaf node represents a class label (in classification) or a value (in regression).

#### How it works:

#### Start at the root node: The algorithm selects the best feature to split the data.

#### Split the data: The dataset is divided into subsets based on the selected feature.

#### Repeat recursively: The process is repeated for each subset until a stopping criterion is met (e.g., maximum depth, minimum samples per leaf).

#### Assign a label/value: Each leaf node is assigned a class label (classification) or a value (regression).



## 2. What are impurity measures in Decision Trees?

####  Impurity measures in Decision Trees are metrics used to evaluate how well a dataset is split at a node. They quantify the homogeneity (or purity) of the target variable within the subsets created by a split. The goal of a Decision Tree is to minimize impurity, leading to more homogeneous subsets.

#### Common Impurity Measures:
#### Gini Impurity: Used in the CART (Classification and Regression Trees) algorithm.
#### Entropy: Used in algorithms like ID3 and C4.5.
#### Variance (for Regression Trees): Used in regression tasks to measure the spread of target values in a subset.

## 5. What is Information Gain, and how is it used in Decision Trees?

#### Information Gain is a metric used in Decision Trees to determine the effectiveness of a feature in splitting the dataset. It measures the reduction in impurity (e.g., Gini Impurity or Entropy) achieved by splitting the data on a particular feature. The feature that provides the highest Information Gain is chosen for the split at each node.

#### Formula:  

#### How is Information Gain Used in Decision Trees: 

#### 1. Compute the Gini Impurity or Entropy of the parent node before splitting.
#### 2. For each feature, calculate the impurity of the child nodes that would result from splitting the data.
#### 3. Use the formula to calculate the Information Gain for each feature.
#### 4. Choose the feature with the highest Information Gain to split the data at the current node.
#### 5. Continue the process for each child node until a stopping criterion is met (e.g., maximum depth, minimum samples per leaf).



## 6. What is the difference between Gini Impurity and Entropy?

#### Gini Impurity: Computationally faster as it doesnt involve logarithms. Ranges from 0 (perfectly pure) to 0.5 (maximally impure for binary classification). More sensitive to changes in class probabilities.

#### Entropy: Involves logarithms, making it computationally slower. Ranges from 0 (perfectly pure) to 1 (maximally impure for binary classification). Tends to produce more balanced trees.

#### In practice, the choice between Gini and Entropy often has little impact on the final model.


## 7. What is the mathematical explanation behind Decision Trees?

####  Decision Trees work by recursively splitting the dataset into subsets based on features to minimize impurity (or maximize homogeneity). Here's how it works:

#### Impurity Measures: Use metrics like Gini Impurity or Entropy to measure how mixed the classes are in a subset. Goal is to Minimize impurity at each step.

#### Splitting: Evaluate all possible splits for each feature. Choose the split that maximizes Information Gain (reduction in impurity).

#### Recursive Partitioning: Repeat the process for each subset until a stopping condition is met (e.g., max depth, minimum samples).

#### Leaf Nodes: For classification, assign the most frequent class in the subset. For regression, assign the average (or median) of the target values.



## 8. What is Pre-Pruning in Decision Trees?

#### Pre-Pruning involves stopping the tree-building process early by setting constraints, such as:

#### Maximum depth of the tree.
#### Minimum number of samples required to split a node.
#### Minimum number of samples required at a leaf node.

#### This prevents the tree from becoming too complex and overfitting the training data.



## 9. What is Post-Pruning in Decision Trees?

#### Post-Pruning involves building the full tree first and then removing branches that do not contribute significantly to the models performance. This is done using techniques like Cost Complexity Pruning (e.g., in CART).



## 10. What is the difference between Pre-Pruning and Post-Pruning?

#### Pre-Pruning: Stops the tree from growing during the training process by setting constraints.
#### Post-Pruning: Builds the full tree first and then prunes it back to improve generalization.

#### Post-Pruning is often more effective but computationally more expensive.



## 11. What is a Decision Tree Regressor?

#### A Decision Tree Regressor is a variant of Decision Trees used for regression tasks. Instead of predicting class labels, it predicts continuous values by averaging the target values in each leaf node.




## 12. What are the advantages and disadvantages of Decision Trees?

#### Advantages: Easy to understand and interpret. Can handle both numerical and categorical data. Requires little data preprocessing (e.g., no need for scaling). Non-parametric (no assumptions about data distribution).

#### Disadvantages: Prone to overfitting, especially with deep trees. Sensitive to small changes in the data. Can create biased trees if some classes dominate.




## 13. How does a Decision Tree handle missing values?

#### Decision Trees can handle missing values by:
#### Using surrogate splits (alternative features to split on).
#### Assigning samples with missing values to the most common branch.
#### Imputing missing values (e.g., using mean/mode).



## 14. How Decision Trees Handle Categorical Features:

#### Direct Splits: Evaluate all ways to split categories (e.g., {Red, Green} vs. {Blue}).

#### Ordinal Encoding: Treat ordered categories as numbers (e.g., Small < Medium < Large).

#### One-Hot Encoding: Convert categories into binary features (e.g., Is_Red, Is_Green, Is_Blue).

#### High Cardinality: Group similar categories or use target encoding.



## 15. What are some real-world applications of Decision Trees?

#### Real-World Applications of Decision Trees:

#### Medical Diagnosis: Predict diseases (e.g., diabetes, cancer).

#### Customer Segmentation: Group customers for targeted marketing.

#### Fraud Detection: Identify fraudulent transactions.

#### Credit Scoring: Assess loan applicants' creditworthiness.

#### Recommendation Systems: Suggest products or movies.

#### Quality Control: Predict defects in manufacturing.

#### Agriculture: Predict crop yields or planting conditions.

#### HR: Screen job applicants or predict employee turnover.

#### Environmental Science: Analyze climate data or species habitats.

## 16. Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy?

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the model
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Calculate and print the model accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

## 21. Write a Python program to train a Decision Tree Classifier with a maximum depth of 3 and compare its accuracy with a fully grown tree.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

# Load dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Decision Tree with max depth of 3
clf_depth_3 = DecisionTreeClassifier(max_depth=3, random_state=42)
clf_depth_3.fit(X_train, y_train)

# Train fully grown Decision Tree
clf_full = DecisionTreeClassifier(random_state=42)
clf_full.fit(X_train, y_train)

# Predict on test data
y_pred_depth_3 = clf_depth_3.predict(X_test)
y_pred_full = clf_full.predict(X_test)

# Calculate accuracy
accuracy_depth_3 = accuracy_score(y_test, y_pred_depth_3)
accuracy_full = accuracy_score(y_test, y_pred_full)

print(f"Accuracy of Decision Tree with max depth 3: {accuracy_depth_3:.4f}")
print(f"Accuracy of Fully Grown Decision Tree: {accuracy_full:.4f}")

# Visualize the decision trees
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plot_tree(clf_depth_3, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.title("Decision Tree (Max Depth = 3)")

plt.subplot(1, 2, 2)
plot_tree(clf_full, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.title("Fully Grown Decision Tree")

plt.show()
```

## 24. Write a Python program to train a Decision Tree Classifier using One-vs-Rest (OvR) strategy for multiclass classification.


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load the Iris dataset (a multiclass classification dataset)
data = load_iris()
X = data.data  # Features
y = data.target  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)

# Wrap the Decision Tree Classifier with OneVsRestClassifier
ovr_classifier = OneVsRestClassifier(decision_tree)

# Train the classifier
ovr_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = ovr_classifier.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Print a detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=data.target_names))
```

## 27. Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
data = load_iris()
X = data.data  # Features
y = data.target  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the classifier without pruning
clf.fit(X_train, y_train)

# Calculate the accuracy of the unpruned tree
y_pred = clf.predict(X_test)
initial_accuracy = accuracy_score(y_test, y_pred)
print(f"Initial Accuracy (Unpruned Tree): {initial_accuracy:.4f}")

# Extract the CCP alphas and impurities from the tree
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas  # Effective alphas for pruning
impurities = path.impurities  # Total impurities at each alpha

# Train classifiers for different ccp_alpha values and store their accuracies
accuracies = []
for ccp_alpha in ccp_alphas:
    clf_pruned = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    clf_pruned.fit(X_train, y_train)
    y_pred_pruned = clf_pruned.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred_pruned))

# Find the optimal ccp_alpha that maximizes accuracy
optimal_alpha = ccp_alphas[np.argmax(accuracies)]
print(f"Optimal ccp_alpha: {optimal_alpha:.4f}")

# Plot the effect of ccp_alpha on accuracy
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, accuracies, marker="o", drawstyle="steps-post")
plt.xlabel("ccp_alpha")
plt.ylabel("Accuracy")
plt.title("Effect of Cost Complexity Pruning on Accuracy")
plt.axvline(optimal_alpha, color="red", linestyle="--", label=f"Optimal ccp_alpha = {optimal_alpha:.4f}")
plt.legend()
plt.grid()
plt.show()

# Train the final pruned tree with the optimal ccp_alpha
final_clf = DecisionTreeClassifier(random_state=42, ccp_alpha=optimal_alpha)
final_clf.fit(X_train, y_train)
y_pred_final = final_clf.predict(X_test)
final_accuracy = accuracy_score(y_test, y_pred_final)
print(f"Final Accuracy (Pruned Tree): {final_accuracy:.4f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report

# Load the Iris dataset
data = load_iris()
X = data.data  # Features
y = data.target  # Labels
class_names = data.target_names  # Class names

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Generate the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Print the classification report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

## 30. Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values for max_depth and min_samples_split.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

# Load the Iris dataset
data = load_iris()
X = data.data  # Features
y = data.target  # Labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Define the parameter grid to search
param_grid = {
    'max_depth': [3, 5, 7, 10, None],  # None means no limit
    'min_samples_split': [2, 5, 10, 15]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, scoring='accuracy')

# Perform the grid search on the training data
grid_search.fit(X_train, y_train)

# Print the best parameters and the corresponding accuracy
best_params = grid_search.best_params_
best_accuracy = grid_search.best_score_
print(f"Best Parameters: {best_params}")
print(f"Best Cross-Validation Accuracy: {best_accuracy:.4f}")

# Train the classifier with the best parameters
best_clf = DecisionTreeClassifier(random_state=42, **best_params)
best_clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = best_clf.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Set Accuracy: {accuracy:.4f}")

# Print a detailed classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=data.target_names))
```
